```python
files = {
    "employee" : "/Volumes/workspace/default/assignment/employees_extended.json",
    "Transaction" : "/Volumes/workspace/default/assignment/transactions_1200 (1).json"
}
df = {}

for name, path in files.items():
    df[name]=spark.read.json(
        path
    )
```

```python
df["Transaction"].display()
```

```python
from pyspark.sql.functions import sum, explode,count,expr,col

df["employee"].select(
    "emp_id","name",explode("projects").alias("project")
).groupBy(
    "emp_id","name"
).agg(
    sum(col("project.hours")).alias("hour"),count(col(
        "project.project_name")).alias("p_name")
    ).filter(
        (col("p_name") > 2) & (col("hour") >100)).display()
```

### Q17: Skill Matching

```python
df["employee"].select("name","emp_id",explode("skills").alias("skill")).filter(col("skill") == "Python").display()
```

### Q18: Monthly Trend

```python
from pyspark.sql.functions import year, month, sum, round, col

df["Transaction"].groupBy(
    year(col("txn_date")).alias("year"),
    month(col("txn_date")).alias("month")
).agg(
    round(
        sum("amount"),
        2
    ).alias("total_amount")
).orderBy(
    "year",
    "month"
).display()
```
