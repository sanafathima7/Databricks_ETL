```python
df = spark.read.csv("/Volumes/workspace/default/assignment/employees (1).csv", header=True, inferSchema=True)
```

### **With Column**

```python
df.withColumn("Salary", df.Salary * 1.10)
```

```python
from pyspark.sql.functions import when

df.withColumn(
    "Category",
    when(df.Salary > 60000, "High")
    .otherwise("Normal")
)
```

## select() → pick columns

```python
df.select("Name", "Salary").show()
```

###  Rename column

```python
df.select(
  df.Name.alias("Employee_name")
).show()
```

### filter() → WHERE

```python
df.filter(df.Department == "IT").show()
```

### Multiple conditions

```python
df.filter(
    (df.Department == "IT") &
    (df.Salary > 50000)
).show()
```

```python
df.filter(
    (df.Department == "IT") |
    (df.Department == "HR")
).show()
```
